```python
%matplotlib inline
import pandas
```

# Covid Data

```python
df = pandas.read_excel('s3://ia241-igc/covid_data.xls')
df[:10]
```

# List of Unique Countries

```python
pandas.unique(df['countriesAndTerritories'])
```

# Somalia Data

```python
Somalia_data= df.loc[df['countriesAndTerritories']=='Somalia']
Somalia_data[:10]
```

# What are the trends of Covid 19 cases in Somalia over time?

```python
covid_cases_over_time = Somalia_data.groupby('dateRep').sum()['cases']
covid_cases_over_time.plot()
```

# What are the deaths based on cases?

```python
Somalia_data.plot.scatter(x='cases',y='deaths')
```

# What are the months with the highest number of covid 19 cases in Somalia?

```python
sum_cases_by_month= Somalia_data.groupby('month').sum()['cases']
sum_cases_by_month.plot.bar()
```
